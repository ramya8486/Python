Preprocessing

It is the process of converting or mapping data from one “raw” form into another
format to make it ready for further analysis.
Data pre-processing is also often called “data cleaning” or “data wrangling”

```python
# To check for missing values
unemployment.isnull().sum()
```

```python
# Drop rows with missing values
unemployment.dropna(inplace=True)
# To check for missing values
unemployment.isnull().sum()
```

```python
# To Check for Duplicates

unemployment.duplicated().sum()
```

```python
# Drop rows with duplicates

unemployment.drop_duplicates(inplace=True)
unemployment.head(10)
```

```python
# To find the unique value

unemployment.nunique()
```

```python
# To drop specific row or column
# dataframe.dropna(subset['column_name'],axis =1, inplace= True)
# axis = 0 drops entire row
# axis =1 drops the entire column

unemployment.drop(['indicator_name'],axis =1, inplace= True)
unemployment
```

Exploratory Data Analysis (EDA)

Exploratory Data Analysis (EDA) is all about examining and visualizing data to grasp its key characteristics, identify trends, and see how various elements relate to each other. 

It’s a great way to catch any odd data points or outliers, and it typically happens before diving into more in-depth statistical analysis or creating models.

```python
print(unemployment.value_counts("age_group").plot.bar())
```

```python
unemployment.value_counts("country_name").head(10).plot.bar()
```

```python
# Unemployment rate distribution in the year of 2024
sns.histplot(unemployment['2024'], bins=20, kde=True)
# Adding title and labels 
plt.title('Unemployment rate distribution in the year 2024') 
plt.xlabel('Unemployment rate %') 
plt.ylabel('count') 
  
# Display the histogram 
plt.show() 
```

```python
# plot the unemployment ratio in the year 2024 based on sex and the age_group

sns.barplot(x='sex', y='2024', hue='age_group', data=unemployment)
# Display plot
plt.show()
```

```python
# Compare unemployment rates across gender and age groups
plt.figure(figsize=(12,6))
sns.barplot(x='sex', y='2024', hue='age_group', data=unemployment)
plt.title('Unemployment Rate by Gender and Age Group in 2024')
plt.xlabel('Gender')
plt.ylabel('Unemployment Rate %')
plt.show()
```

```python

# Create pair plot for comparing the unemployment ratio for the years 2023,2024

unemploy=unemployment.pivot( columns='sex', values=["2023","2024"])
sns.pairplot(data=unemploy)

# Show plot
plt.show()
```

```python
# Average unemployment rate by year
avg_unemployment_year = unemployment[['2014', '2015', '2016', '2016', '2017', '2018', '2019', '2020', '2021', '2022', '2023', '2024']].mean()
plt.figure(figsize=(20,8))
avg_unemployment_year.plot(kind='line', marker='o')
plt.title('Average Unemployment Rate Over the Years')
plt.xlabel('Year')
plt.ylabel('Average Unemployment Rate %')
plt.show()
```

```python

# Exclude non-numeric columns for correlation analysis
numeric_columns = unemployment.select_dtypes(include='number').columns
correlation_matrix = unemployment[numeric_columns].corr()
# Plot the correlation heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```
